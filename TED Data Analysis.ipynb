### 2. Whick talks provoke the most online discussion

```python
# Sort by the number of comments
ted.sort_values('comments').tail()
```

```python
# Calculate number of comments per view
ted["comments_per_view"] = ted.comments / ted.views
```

```python
ted.sort_values("comments_per_view").tail()
```

```python
#Calculate no of views per comment - reverse the previous calculation
ted["views_per_comment"] = ted.views / ted.comments
ted.sort_values("views_per_comment").tail(3)
```

```python
ted.sort_values("views_per_comment").head(3)
# interpertation - one out of every 450 people leave a comment
```
